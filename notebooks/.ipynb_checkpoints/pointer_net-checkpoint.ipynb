```python
import pandas as pd
import os
```

```python
df = pd.read_csv('data/bangla_chakma.csv').dropna().sample(frac=1)
df
```

```python
def split(s):
  s = s.replace(" ", "_")
  s = ' '.join(list(s))
  return s
```

```python
def merge(s):
  s = s.replace('\n', '_')
  s = ''.join(s.split())
  s = s.replace("_", ' ')
  return s
```

```python
df['Bangla'] = df['Bangla'].apply(lambda x: split(x))
df['Chakma'] = df['Chakma'].apply(lambda x: split(x))
df
```

```python
data = {
    "train": df,
    "test": df.sample(1000),
    "valid": df.sample(1000)
}
```

```python
!rm data/*.txt
```

```python
for key in data.keys():
  fp_input = open("data/src_{}.txt".format(key), 'w+')
  fp_input.write('\n'.join(list(data[key]['Bangla'].values)))
  fp_input.close

  fp_input = open("data/tgt_{}.txt".format(key), 'w+')
  fp_input.write('\n'.join(list(data[key]['Chakma'].values)))
  fp_input.close
```

```python
import subprocess

def tranlsate(NL):
  nl_file = open("single_test.txt", "w")
  nl_file.write(NL)
  nl_file.close()
  s = "python OpenNMT-py-legacy/translate.py -model {} -src single_test.txt -output sar_to_compile.txt -replace_unk -verbose -beam_size 1".format(best_model_dir)
  subprocess.call(s.split())
  f = open("sar_to_compile.txt", 'r')
  sar = f.read().strip()
  f.close()
  subprocess.call('rm single_test.txt'.split())
  subprocess.call('rm sar_to_compile.txt'.split())
  return sar
```

```python
NL = "create an app with a button , a textbox , and a text2speech . when the button is pressed , speak the textbox text ."
tranlsate(NL)
```
